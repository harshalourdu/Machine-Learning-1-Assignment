# Q1: Explain the following with an example:
C) Artificial Intelligence (AI)
Artificial Intelligence refers to the simulation of human intelligence in machines that are programmed to think and learn like humans. AI involves creating systems that can perform tasks such as problem-solving, reasoning, and decision-making without human intervention.

Example: A chatbot, such as Siri or Alexa, is an example of AI. It understands natural language, responds to queries, and learns from interactions to improve its performance.
D) Machine Learning (ML)
Machine Learning is a subset of AI that focuses on building algorithms that enable systems to learn from data and make decisions based on that data. It involves training a model using a set of data and using it to make predictions or decisions without being explicitly programmed for each task.

Example: A spam email filter is an example of machine learning. It is trained on a dataset of emails marked as spam or not spam, and over time it learns to classify new emails as either spam or not based on the features learned from the training data.
E) Deep Learning (DL)
Deep Learning is a subset of machine learning that uses neural networks with many layers (hence "deep") to model complex patterns in large amounts of data. It is particularly powerful for tasks like image recognition, speech recognition, and natural language processing.

Example: A self-driving car uses deep learning to analyze images from cameras and make decisions about navigating the road by recognizing objects, other cars, pedestrians, and traffic signs.



# Q2: What is supervised learning? List some examples of supervised learning.
Supervised learning is a type of machine learning where the model is trained on labeled data. Each input in the training dataset is paired with the correct output (or label), and the model learns to map the inputs to the correct outputs.

Examples of Supervised Learning:
Linear Regression: Predicting house prices based on features like size, location, and number of rooms.
Logistic Regression: Classifying whether an email is spam or not.
Decision Trees: Classifying types of fruits based on attributes like color, size, and texture.
Support Vector Machines (SVM): Classifying emails as spam or not spam.
K-Nearest Neighbors (KNN): Classifying customers based on past purchase behavior.


# Q3: What is unsupervised learning? List some examples of unsupervised learning.
Unsupervised learning involves training a model on data that has no labels. The goal is to find hidden patterns, structures, or relationships within the data without predefined outcomes.

Examples of Unsupervised Learning:
Clustering:
K-Means: Grouping customers into clusters based on purchasing behavior.
Hierarchical Clustering: Segmenting different types of customers based on attributes like age, income, and buying preferences.
Dimensionality Reduction:
Principal Component Analysis (PCA): Reducing the number of features in a dataset while maintaining the variance, such as in face recognition.
Anomaly Detection: Identifying rare events or outliers, such as detecting fraudulent transactions in banking.
Autoencoders: Used for feature learning and data compression in neural networks.



# Q4: What is the difference between AI, ML, DL, and DS?
Artificial Intelligence (AI): The broader concept of machines being able to perform tasks that would typically require human intelligence, such as reasoning, problem-solving, and decision-making.

Machine Learning (ML): A subset of AI that involves algorithms that allow computers to learn from data and improve over time without being explicitly programmed.

Deep Learning (DL): A subset of ML that uses deep neural networks with many layers to model complex patterns in large datasets, often used in tasks such as image and speech recognition.

Data Science (DS): Data Science involves using scientific methods, processes, algorithms, and systems to extract insights and knowledge from structured and unstructured data. It often involves ML and AI techniques to analyze data and make predictions or decisions.



# Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
Supervised Learning:

Data: Requires labeled data (input-output pairs).
Purpose: The goal is to learn a mapping from inputs to outputs (e.g., classification, regression).
Example: Predicting house prices based on features like size and location.
Unsupervised Learning:

Data: Uses unlabeled data (only inputs, no outputs).
Purpose: The goal is to find hidden structures or patterns in the data (e.g., clustering, anomaly detection).
Example: Grouping customers into clusters based on purchasing behavior.
Semi-supervised Learning:

Data: Uses a small amount of labeled data and a large amount of unlabeled data.
Purpose: It bridges the gap between supervised and unsupervised learning and is useful when labeling data is expensive or time-consuming.
Example: Classifying images where only a few images are labeled, and the rest are unlabeled.



# Q6: What is train, test, and validation split? Explain the importance of each term.
Train Set: The portion of the data used to train the machine learning model. The model learns from this data by adjusting its internal parameters.
Importance: Provides the model with examples to learn from and make predictions.
Test Set: A separate portion of the data used to evaluate the model's performance after training. It provides an unbiased estimate of the model's generalization ability.
Importance: Helps assess how well the model will perform on unseen data.
Validation Set: A separate portion of the data used during the training process to tune hyperparameters and avoid overfitting. The validation set is used to adjust the model before the final evaluation on the test set.
Importance: Provides a method to optimize the model's performance and reduce overfitting.



# Q7: How can unsupervised learning be used in anomaly detection?
Unsupervised learning is effective in anomaly detection because it can identify patterns and outliers in data that do not conform to expected behaviors. Since unsupervised learning does not require labeled data, it can automatically find anomalies based on the distribution and relationships within the data.

Example: In fraud detection, unsupervised learning can be used to find unusual spending patterns in credit card transactions, which could indicate fraudulent activity. Common techniques for anomaly detection include clustering methods like K-Means and density-based methods like DBSCAN.
